```python
clientes = pd.read_parquet("datasets/clientes.parquet")
productos = pd.read_parquet("datasets/productos.parquet")
transacciones = pd.read_parquet("datasets/transacciones.parquet")
```

```python
print(clientes["customer_id"].duplicated().sum())
print(productos["product_id"].duplicated().sum())
print(transacciones.duplicated().sum())
```

Se observa que hay 885 filas duplicadas en transacciones, por lo que se eliminan a continuaci√≥n.

```python
transacciones= transacciones.drop_duplicates()
```

```python
print("Clientes:")
print(clientes.dtypes)
print("----")
print("Productos:")
print(productos.dtypes)
print("----")
print("Transacciones:")
print(transacciones.dtypes)
```

Se observa la existencia de muchos par√°metros del tipo object, por lo que por estandarizaci√≥n, se prefieren pasar a tipo string.

```python
clientes["customer_type"] = clientes["customer_type"].astype("string")
for col in ["brand", "category", "sub_category", "segment", "package"]:
    productos[col] = productos[col].astype("string")
```

Finalmente, al observar la descripci√≥n de las columnas, la mayor√≠a podr√≠an tener alguna relevancia, la cual puede ser estudiada en el eda, sin embargo la latitud y longitud geogr√°fica ("X","Y") pierden toda relevancia si existen par√°metros m√°s claros como "zone_id" y "region_id". Por lo mismo se decide borrarlas previo a hacer an√°lisis exploratorio.


```python
clientes.drop(["X", "Y"], axis=1, inplace=True)
```

Finalmente, para poder hacer una tabla con todos los datos para poder trabajar con Machine learning, hacemos un doble merge usando como base la tabla de transacciones.

```python
data = transacciones.merge(clientes, on="customer_id", how="left")
data = data.merge(productos, on="product_id", how="left")
```

Se puede estandarizar el texto del dataset, para que no haya problemas en la  escritura de las categor√≠as en el set de datos.

```python
for c in ["customer_type","brand","category","sub_category","segment","package"]:
    data[c] = data[c].astype("string").str.strip().str.lower().str.replace(r"\s+", " ", regex=True)
```

Finalmente, al observar los dtype de items es float, pero esto no calza con la definici√≥n de cantidad de bultos vendido , al no poder existir decimales en este.

Por lo mismo se trnasforma a int, haciendo aproximaci√≥n al natural m√°s cercano.


```python
data["items"] = np.floor(data["items"] + 0.5).astype("Int64")
```

```python
# Transacciones por d√≠a.
for mes in range(1, 13):
    df_mes = data[data["purchase_date"].dt.month == mes]

    frecuencia_diaria = df_mes.groupby(df_mes["purchase_date"].dt.date).size()

    plt.figure(figsize=(12,6))
    frecuencia_diaria.plot(kind="bar", color="skyblue", edgecolor="black")
    plt.title(f"Frecuencia de transacciones por d√≠a - Mes {mes}")
    plt.xlabel("D√≠a")
    plt.ylabel("Cantidad de transacciones")
    plt.xticks(rotation=45)
    plt.grid(axis="y", alpha=0.4)
    plt.show()
```

En este caso, se observa como la frecuencia de los registros es diaria, al haber entradas de datos todos los d√≠as del a√±o. Al hacer los gr√°ficos, se puede observar que en general todos los meses tienen un comportamiento similar, con meses con m√°s ventas en general (Por ejemplo, Diciembre).
Adem√°s, se observan ciertos peaks de compra en cada mes.

```python
#Transacciones por d√≠a.
diario = data.groupby(data["purchase_date"].dt.normalize()).size()
prom = diario.groupby(diario.index.dayofweek).mean().reindex(range(7))
dias = ["Lunes","Martes","Mi√©rcoles","Jueves","Viernes","S√°bado","Domingo"]

plt.figure(figsize=(9,4))
bars = plt.bar(dias, prom.values, edgecolor="black")
plt.title("Promedio de transacciones por d√≠a de la semana")
plt.xlabel("D√≠a"); plt.ylabel("Promedio de transacciones por d√≠a")
plt.grid(axis="y", alpha=0.3)


for b, v in zip(bars, prom.values):
    plt.text(b.get_x()+b.get_width()/2, b.get_height()*1.01, f"{v:.1f}",
             ha="center", va="bottom", fontweight="bold", fontsize=6.5)

plt.tight_layout()
plt.show()
```

Si agrupamos las ventas por d√≠a, se observa una estacionalidad semanal muy marcada. Donde en general,los fines de semana la cantidad de transacciones disminuye fuertemente.

El objetivo es aplicar una serie de transformaciones en un orden definido para asegurar que los datos est√©n listos para los modelos a entrenar. El pipeline final debe incluir las t√©cnicas de pre-procesamiento que se deben aplicar a los distintos datos (seg√∫n lo que consideren necesario para el problema). Por ejemplo:

- **Imputaci√≥n de valores nulos**: Manejo de datos faltantes mediante estrategias adecuadas (media, mediana, moda, interpolaci√≥n, etc.).

- **Transformaciones personalizadas**: Uso de ColumnTransformer para aplicar diferentes transformaciones a columnas espec√≠ficas.

- **Codificaci√≥n de variables categ√≥ricas**: Convertir datos categ√≥ricos a un formato num√©rico adecuado (One-Hot Encoding, Label Encoding, etc.).

- **Discretizaci√≥n de variables**: Conversi√≥n de variables num√©ricas continuas en categor√≠as si son relevantes para el desempe√±o del modelo a entrenar.

- **Estandarizaci√≥n o normalizaci√≥n** : Ajustar la escala de los datos para mejorar el rendimiento de los algoritmos sensibles a la magnitud de las variables.

- **Eliminaci√≥n o transformaci√≥n de valores at√≠picos**: Identificar y tratar con datos outliers para mejorar la robustez del modelo.

- **Nuevas caracter√≠sticas** : Creaci√≥n de variables adicionales que puedan aportar informaci√≥n relevante al modelo.

Cada una de estas transformaciones debe ser justificada en funci√≥n de su relevancia para el problema y los datos, y es importante evaluar su impacto en el rendimiento del modelo. Adem√°s, el pipeline debe ser flexible y modular para poder probar diferentes configuraciones de preprocesamiento.

Inicialmente, se generan las clases necesarias para el pipeline.

```python
class IQR(BaseEstimator, TransformerMixin):
    def __init__(self, lambd=1.5):
        self.lambd = lambd
    def fit(self, X, y=None):
        return self
    def transform(self, X):
        X = X.copy()
        n_cols = X.select_dtypes(include=[np.number]).columns
        for col in n_cols:
            Q1 = X[col].quantile(0.25); Q3 = X[col].quantile(0.75)
            IQR = Q3 - Q1
            inf = Q1 - self.lambd * IQR; sup = Q3 + self.lambd * IQR
            X.loc[(X[col] < inf) | (X[col] > sup), col] = np.nan
        return X
    def set_output(self, transform='default'):
        return self
```

```python
class MinMax(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        self.min_ = X.min(); self.max_ = X.max()
        return self
    def transform(self, X):
        X = X.copy()
        return (X - self.min_) / (self.max_ - self.min_)
    def set_output(self, transform='default'):
        return self
```

```python
class Date(BaseEstimator, TransformerMixin):
    def __init__(self, col="purchase_date"):
        self.col = col

    def fit(self, X, y=None):
        return self

    def transform(self, X):
        X = X.copy()
        dt = pd.to_datetime(X[self.col])
        X["D√≠a"] = dt.dt.dayofweek.astype("Int64")      # 0=lunes ‚Ä¶ 6=domingo
        X["Mes"] = dt.dt.month.astype("Int64")
        X["Fds"] = dt.dt.dayofweek.isin([5,6]).astype("Int64")
        return X

    def set_output(self, transform='default'):
        return self
```

```python
class Negative(BaseEstimator, TransformerMixin):
    """Reemplaza valores < 0 por NaN en las columnas entregadas a este transformer."""
    def fit(self, X, y=None):
        return self
    def transform(self, X):
        X = X.copy()
        num_cols = X.select_dtypes(include=[np.number]).columns
        for c in num_cols:
            X.loc[X[c] < 0, c] = np.nan
        return X
    def set_output(self, transform='default'):
        return self
```
