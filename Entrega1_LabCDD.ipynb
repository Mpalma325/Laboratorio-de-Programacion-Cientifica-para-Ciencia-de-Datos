```python
#Patrones de compra
# Crear columna de aﾃｱo-semana
data['year_week'] = data['purchase_date'].dt.strftime('%Y-%U')

# Calcular productos comprados por cliente y semana
productos_por_semana = data.groupby(['customer_id', 'year_week'])['product_id'].nunique().reset_index(name='productos en semana')

# Promedio de productos por cliente (en todas sus semanas activas)
avg_items_week = productos_por_semana.groupby('customer_id')['productos en semana'].mean().reset_index(name='avg_items_per_week')

avg_items_week
```

```python
mx = avg_items_week['avg_items_per_week'].max()
step = 10
top = int(np.ceil((mx + 1) / step) * step)

bins = list(range(0, top + 1, step))
labels = [f"[{bins[i]},{bins[i+1]})" for i in range(len(bins)-1)]

avg_items_week['avg_range'] = pd.cut(
    avg_items_week['avg_items_per_week'],
    bins=bins,
    right=False,
    labels=labels,
    include_lowest=True
)

counts_avg = (
    avg_items_week['avg_range']
    .value_counts()
    .sort_index()
    .reset_index(name='num_clients')
    .rename(columns={'index': 'avg_range'})
)

counts_avg
```

```python
plt.figure(figsize=(12,6))
plt.bar(counts_avg['avg_range'].astype(str), counts_avg['num_clients'],
        color="lightgreen", edgecolor="black")

plt.xticks(rotation=90)
plt.xlabel("Rango de ﾃｭtems promedio por semana")
plt.ylabel("Nﾃｺmero de clientes")
plt.title("Distribuciﾃｳn de clientes por promedio semanal de ﾃｭtems")
plt.tight_layout()
plt.show()
```

A partir del grafico, donde se expone el promedio semanal de ﾃｭtems por cantida de clientes, se evidencia que la gran mayorﾃｭa de los clientes compra en promedio entre 0 y 10 ﾃｭtems por semana, lo que indica que el comportamiento tﾃｭpico es de bajo volumen semanal.

Luego, a medida que aumenta el nﾃｺmero de transacciones, el nﾃｺmero de clientes disminuye de forma progresiva.

```python
# Ordenar por producto y fecha
data = data.sort_values(by=['product_id', 'purchase_date'])

# Diferencia en dﾃｭas entre compras sucesivas del mismo SKU
data['days_diff'] = data.groupby('product_id')['purchase_date'].diff().dt.days

# Promedio de recompra (en dﾃｭas) por SKU
avg_repurchase_sku = data.groupby('product_id')['days_diff'].mean()

avg_repurchase_sku.head()
```

```python
avg_repurchase_sku = avg_repurchase_sku.reset_index(name='avg_days_diff')
mx = avg_repurchase_sku['avg_days_diff'].max()
step = 10
top = int(np.ceil((mx + 1) / step) * step)

bins = list(range(0, top + 1, step))
labels = [f"[{bins[i]},{bins[i+1]})" for i in range(len(bins)-1)]

avg_repurchase_sku['days_range'] = pd.cut(
    avg_repurchase_sku['avg_days_diff'],
    bins=bins,
    right=False,
    labels=labels,
    include_lowest=True
)

counts_repurchase = (
    avg_repurchase_sku['days_range']
    .value_counts()
    .sort_index()
    .reset_index(name='num_skus')
    .rename(columns={'index': 'days_range'})
)

counts_repurchase
```

```python
plt.figure(figsize=(12,6))
plt.bar(counts_repurchase['days_range'].astype(str), counts_repurchase['num_skus'],
        color="salmon", edgecolor="black")

plt.xticks(rotation=90)
plt.xlabel("Rango de promedio de recompra (dﾃｭas)")
plt.ylabel("Nﾃｺmero de SKUs")
plt.title("Distribuciﾃｳn de SKUs por promedio de recompra")
plt.tight_layout()
plt.show()
```

A partir del grafico, donde se expone el promedio de recompra por SKU (realizado a partir del ID de los productos), se evidencia que casi todos los SKU's tienen un promedio de recompra muy bajo (menos de 10 dﾃｭas) y existen algunos SKUs con ciclos de recompra mﾃ｡s largos, pero son casos aislados.

A partir de los 3 graficos expuestos anteriormente, asociados a los patrones de compra, se evidencia que, en promedio, los clientes suelen comprar pocos ﾃｭtems distintos por semana, realizando, mayormente, entre 20 y 40 transacciones en el periodo, mientras que los SKUs tienen una recompra muy rﾃ｡pida, lo que sugiere que la mayoria son productos de alta rotaciﾃｳn (o bien, de "consumo rapido").
