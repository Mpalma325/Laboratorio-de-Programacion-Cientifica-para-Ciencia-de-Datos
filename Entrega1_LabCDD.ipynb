```python
# Transacciones por d√≠a.
for mes in range(1, 13):
    df_mes = data[data["purchase_date"].dt.month == mes]

    frecuencia_diaria = df_mes.groupby(df_mes["purchase_date"].dt.date).size()

    plt.figure(figsize=(12,6))
    frecuencia_diaria.plot(kind="bar", color="skyblue", edgecolor="black")
    plt.title(f"Frecuencia de transacciones por d√≠a - Mes {mes}")
    plt.xlabel("D√≠a")
    plt.ylabel("Cantidad de transacciones")
    plt.xticks(rotation=45)
    plt.grid(axis="y", alpha=0.4)
    plt.show()
```
