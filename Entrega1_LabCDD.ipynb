```python
#Patrones de compra
# Crear columna de año-semana
data['year_week'] = data['purchase_date'].dt.strftime('%Y-%U')

# Calcular productos comprados por cliente y semana
productos_por_semana = data.groupby(['customer_id', 'year_week'])['product_id'].nunique().reset_index(name='productos en semana')

# Promedio de productos por cliente (en todas sus semanas activas)
avg_items_week = productos_por_semana.groupby('customer_id')['productos en semana'].mean().reset_index(name='avg_items_per_week')

avg_items_week
```

```python
mx = avg_items_week['avg_items_per_week'].max()
step = 10
top = int(np.ceil((mx + 1) / step) * step)

bins = list(range(0, top + 1, step))
labels = [f"[{bins[i]},{bins[i+1]})" for i in range(len(bins)-1)]

avg_items_week['avg_range'] = pd.cut(
    avg_items_week['avg_items_per_week'],
    bins=bins,
    right=False,
    labels=labels,
    include_lowest=True
)

counts_avg = (
    avg_items_week['avg_range']
    .value_counts()
    .sort_index()
    .reset_index(name='num_clients')
    .rename(columns={'index': 'avg_range'})
)

counts_avg
```

```python
plt.figure(figsize=(12,6))
plt.bar(counts_avg['avg_range'].astype(str), counts_avg['num_clients'],
        color="lightgreen", edgecolor="black")

plt.xticks(rotation=90)
plt.xlabel("Rango de ítems promedio por semana")
plt.ylabel("Número de clientes")
plt.title("Distribución de clientes por promedio semanal de ítems")
plt.tight_layout()
plt.show()
```

A partir del grafico, donde se expone el promedio semanal de ítems por cantida de clientes, se evidencia que la gran mayoría de los clientes compra en promedio entre 0 y 10 ítems por semana, lo que indica que el comportamiento típico es de bajo volumen semanal.

```python
# Ordenar por producto y fecha
data = data.sort_values(by=['product_id', 'purchase_date'])

# Diferencia en días entre compras sucesivas del mismo SKU
data['days_diff'] = data.groupby('product_id')['purchase_date'].diff().dt.days

# Promedio de recompra (en días) por SKU
avg_repurchase_sku = data.groupby('product_id')['days_diff'].mean()

avg_repurchase_sku.head()
```

## 📌 Holdout [0.25 puntos]

Para evaluar correctamente el modelo y garantizar su capacidad de generalización, se deben dividir los datos en tres conjuntos:
- `Entrenamiento` : Para ajustar los parámetros.
- `Validación`: Para optimizar hiperparámetros y seleccionar el mejor modelo.
- `Prueba` : Para evaluar el rendimiento final en datos no vistos.

👀 **Hint**: *Recuerde que los datos tienen una temporalidad que debe considerarse al momento de separarlos, para evitar fugas de información. Es importante justificar la estrategia de partición elegida y visualizar la distribución temporal de los conjuntos generados*

```python
from sklearn.model_selection import train_test_split
import matplotlib.dates as mdates

data = data.sort_values(by='purchase_date')

train_size = 0.7
val_size = 0.15
test_size = 0.15

train_end = int(len(data) * train_size)
val_end = int(len(data) * (train_size + val_size))

train_data = data.iloc[:train_end]
val_data = data.iloc[train_end:val_end]
test_data = data.iloc[val_end:]

print(f"Entrenamiento: {train_data.shape}, Fechas: {train_data['purchase_date'].min()} a {train_data['purchase_date'].max()}")
print(f"Validacion: {val_data.shape}, Fechas: {val_data['purchase_date'].min()} a {val_data['purchase_date'].max()}")
print(f"Test: {test_data.shape}, Fechas: {test_data['purchase_date'].min()} a {test_data['purchase_date'].max()}")

fig, ax = plt.subplots(figsize=(12, 6))
ax.hist(train_data['purchase_date'], bins='auto', alpha=0.7, label='Train')
ax.hist(val_data['purchase_date'], bins='auto', alpha=0.7, label='Validation')
ax.hist(test_data['purchase_date'], bins='auto', alpha=0.7, label='Test')
ax.xaxis.set_major_locator(mdates.MonthLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
plt.xticks(rotation=45)
plt.xlabel('Purchase Date')
plt.ylabel('Number of Transactions')
plt.title('Temporal Distribution of Data Splits')
plt.legend()
plt.show()
```
